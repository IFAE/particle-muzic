```python
import math
import pypdt
import numpy as np
from __future__ import division
tbl = pypdt.ParticleDataTable()

part_dict = {}
for p in tbl:
    part_dict[p.name]= p.id

def magnitude(x):
    return int(math.log10(x))

lifetime_array = []
lifetime_mag = []
lifetime_ren = []
lifetime_dict = {}
mass_array = []
mass_dict = {}
for p in tbl: 
    if p.lifetime != None:
        mass_array.append(p.mass)
        mass_dict[p.name]=p.mass
        if p.lifetime != None:
            lifetime_array.append(p.lifetime)
            lifetime_mag.append(magnitude(p.lifetime))
            lifetime_dict[p.name]=magnitude(p.lifetime)
            lifetime_ren.append(renormalize_time(magnitude(p.lifetime),range1,range2))
        
            
lifetime_array=np.asarray(lifetime_array)
lifetime_mag=np.asarray(lifetime_mag)
lifetime_ren=np.asarray(lifetime_ren)
mass_array=np.asarray(mass_array)


def search_mag(mag):
    for name, magnitude in lifetime_dict.iteritems():
        if magnitude == mag:
            print name
    
def search_mass(mass):
    for name, m in mass_dict.iteritems():
        if m == mass:
            print name

def renormalize_time(n, range1, range2):
    range1=[-13,14]
    range2=[1e-3,5]
    delta1 = range1[1] - range1[0]
    delta2 = range2[1] - range2[0]
    return np.around((delta2 * (n - range1[0]) / delta1) + range2[0],2)

```
