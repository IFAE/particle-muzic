```python
import math
from __future__ import division

poly = []
n = 100
n*=1

w = 640
h = 640
def renormalize(value,start1,end1,start2,end2):
    delta1 = end1 - start1
    delta2 = end2 - start2
    return (delta2*(value - start1)/delta1)+start2
for i in range(0,n):
    a = {}
    a['x']= w/2 + 100*math.sin(renormalize(i, 0, n-1, 0, 2*math.pi))
    a['y']= h/2 + 100*math.cos(renormalize(i, 0, n-1, 0, 2*math.pi))
    poly.append(a)
```
